## Data Visualisation on "AIRBNB DATASET"

```python
area=airbnb['neighbourhood_group']
review=airbnb['number_of_reviews']

plt.bar(area,review,color='pink', width=0.8)
plt.title('Area wise Review',fontsize=12)
plt.xlabel('AREA',fontsize=12)
plt.ylabel('REVIEWS',fontsize=12)
plt.show
```

### By using the 'neighbourhood_group' and 'number_of_reviews' column from the AIRBNB dataset, we are looking at the area wise reviews received using Barplot

```python
price=airbnb['price']
review=airbnb['number_of_reviews']

plt.scatter(price,review)
plt.title('Price Vs Reviews',fontsize=12)
plt.xlabel('PRICE',fontsize=12)
plt.ylabel('REVIEWS',fontsize=12)
plt.show
```

### By using the 'price' and 'number_of_reviews' column from the AIRBNB dataset, we are looking at the price charged and the number of reviews received using Scatterplot and we can see that people prefer to pay lesser price

# SECOND DATASET USED IS "HR DATASET"

```python
HRdata = pd.read_csv(r"C:\Users\Gurdeep\Desktop\DATA SCIENTIST\DATA IS GOOD\Internship Unmessenger\Assignments\Assignment 4_Data Vizualization using Python Libraries\HR Dataset.csv")
```

```python
HRdata.head()
```

```python
HRdata.columns
```

```python
HRdata.drop(['EmpID','MarriedID','MaritalStatusID','GenderID','EmpStatusID','DeptID','FromDiversityJobFairID','Termd','PositionID','Zip','DOB','HispanicLatino','ManagerID','ManagerName','LastPerformanceReview_Date','DaysLateLast30'],axis=1,inplace=True)
```

```python
HRdata.head()
```

```python
HRdata['Sex'].replace('M ', 'Male', inplace= True)
HRdata['Sex'].replace('F', 'Female', inplace= True)
```

```python
HRdata.head()
```

```python
HRdata['DateofTermination']= pd.to_datetime(HRdata['DateofTermination'])
```

```python
HRdata.dtypes
```

```python
HRdata.isna().sum()
```

```python
HRdata.head()
```

## Data Visualisation on "HR DATASET"

```python
sns.jointplot(x='Sex',y='Salary', data= HRdata)
plt.show
```

### By using the 'Sex' and 'Salary' column from the HR dataset, we are looking at the salary received by which gender using Jointplot

##### 

```python
male=HRdata.loc[HRdata['Sex']=='Male'].count()[0]
female=HRdata.loc[HRdata['Sex']=='Female'].count()[0]

labels=['male','female']
colors=['Blue','Pink']

plt.title('Number of Males/Females working',fontsize=12)

plt.pie([male,female],labels=labels,colors=colors,pctdistance=0.8, autopct='%.2f %%')

plt.show
```

### By using the 'Sex' column from the HR dataset, we are looking at the percentage of males and female working by using Pie plot

##### 

```python
plt.figure(figsize=(20,8))
sns.violinplot(data=HRdata, x='Department', y='PerfScoreID')
plt.title('Performance Score by Department', fontsize=12)
plt.xlabel('Department', fontsize=12)
plt.ylabel('Performance Score', fontsize=12)
plt.show
```

### By using the 'Department' and 'PerfScoreID' column from the HR dataset, we are comparing the department wise performance using Violinplot

##### 

```python
plt.figure(figsize=(20, 8))
sns.countplot(data=HRdata, x='Department', hue='EmploymentStatus')
plt.xlabel('Department')
plt.ylabel('Count')
plt.title('Employment Status by Department')
plt.legend(title='Employment Status', loc='upper right')
plt.xticks(rotation=45)
plt.show
```

### By using the 'Department' and 'EmploymentStatus' column from the HR dataset, we are looking at the department wise employment status using countplot

##### 

```python
data = HRdata[['RecruitmentSource','Position','EngagementSurvey']].head()
data.plot.barh(x='RecruitmentSource',title = "Recruitement Source used",figsize=(15,5), legend = True,fontsize=16)
plt.xlabel('Engagement Survey',fontsize=12)
plt.ylabel('Recruitment Source',fontsize=12)
plt.show
```

### By using the 'RecruitmentSource','Position' and 'EngagementSurvey' column from the HR dataset, we are checking for the best recruitment source used which has given the most engaging employees using Barh plot

##### 

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='EmpSatisfaction', y='Absences',hue='Sex', data=HRdata)
plt.legend(title='Gender', loc='upper left')
plt.show
```

### By using the 'EmpSatisfaction', 'Absences' and 'Sex' column from the HR dataset, we can see how many employees are satisfied and take less leaves based on their gender using Box Plot
